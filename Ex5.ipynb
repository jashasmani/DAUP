```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Details.csv')
```

```python
x = np.array(df['Sem 5']).reshape(-1,1)
y = np.array(df['Sem 6']).reshape(-1,1)

plt.scatter(x,y)
plt.colorbar()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
r=LinearRegression()
r.fit(x_train,y_train)
```

```python
y_predict = r.predict(x_test)
plt.scatter(x_test,y_test,color="black",label="original data")
plt.plot(x_test, y_predict, color = "blue", linewidth=3, label = "regression line")
plt.xlabel("Independent Feature")
plt.ylabel("Target Values")
plt.title("Simple Linear Regression")
plt.show()
```
