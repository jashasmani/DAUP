```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

#Kmeans
X, y_true = make_blobs(n_samples=100, centers=3, cluster_std=0.6)
plt.scatter(X[:,0],X[:,1], s=30)  #all rows of 0th column, all rows of 1st column and size of the dots in graph

km = KMeans(n_clusters=3)

km.fit(X)
y_lab = km.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_lab, s=50, cmap='plasma')
centers = km.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);


#DBScan
from sklearn.cluster import dbscan
dbs = dbscan(X)
#print(dbs)
#dbs[0]: Cluster labels. Points labeled -1 are considered as noise.
#dbs[1]: Core sample indices. These are the indices of the points that are core samples (i.e., not noise).

plt.scatter(X[:, 0], X[:, 1], c=dbs[1], s=50, cmap='plasma')
```
