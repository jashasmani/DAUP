```python
# 1. Generate Random Data for clustering
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
X, y_true = make_blobs(n_samples=100, centers=3, cluster_std=0.5)
plt.scatter(X[:,0],X[:,1], s=50)
```

```python
# 2. Create DBSCAN model
from sklearn.cluster import dbscan
dbs = dbscan(X)
```

```python
# 3. Plot Data Points
plt.scatter(X[:, 0], X[:, 1], c=dbs[1], s=50, cmap='viridis')
```

```python
from sklearn import datasets
from sklearn.cluster import DBSCAN

iris=datasets.load_iris()

X=iris.data

dbscan = DBSCAN(eps=0.5,min_samples=10)

y_db=dbscan.fit_predict(X)

plt.scatter(X[:,0],X[:,1],c=y_db)
```

```python
from sklearn import datasets
from sklearn.cluster import DBSCAN
import pandas as pd

df=pd.read_csv('/content/weather.csv')

df['play']=df['play'].replace(to_replace='yes',value=1)
df['play']=df['play'].replace(to_replace='no',value=0)
df['temperature']=df['temperature'].replace(to_replace='hot',value=0.8)
df['temperature']=df['temperature'].replace(to_replace='cool',value=0)
df['temperature']=df['temperature'].replace(to_replace='mild',value=0.5)

X=df[['play','temperature']]
print(X)

dbscan = DBSCAN(eps=0.5,min_samples=10)

y_db=dbscan.fit_predict(X)

plt.scatter(X['play'], X['temperature'], c=y_db)
plt.xlabel('Play')
plt.ylabel('Windy')
plt.show()
```
