```python
import pandas as pd

details=pd.read_csv('/content/sample_data/Details.csv')
print(details)
```

```python
placement= pd.read_csv('/content/sample_data/Placement.csv')
print(placement)
```

```python
all = pd.merge(details,placement,on='Roll No.')
print(all)
```

```python
placement=all[all['Placed']=='Yes'].copy()
placement
```

```python
import random as rd

salary = []
for i in range(len(placement)):
  cpi=placement.loc[i].at['CPI']
  if cpi>=8:
    salary.append(rd.randint(65000,70000))
  elif cpi >= 7 :
    salary.append(rd.randint(60000,65000))
  else :
    salary.append(rd.randint(55000,60000))

placement.insert(12,"Salary",salary,True)
```

```python
placement
```

```python
placement['CPI']=placement['CPI'].fillna(0)
avg=placement['CPI'].mean()
placement['CPI']=placement['CPI'].replace(to_replace=0,value=avg)
placement
```

```python
from datetime import datetime

BirthYear=datetime.now().year-placement['Age']
placement.insert(15,'BirthYear',BirthYear,True)
placement
```

```python
gp=placement.groupby('Branch')
print(gp.get_group('BCA'))
```
